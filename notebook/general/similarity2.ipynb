```python
import RNA
from src.dataset import load_benchmark_dataset, Species, Modification
```

```python
test_dataset = load_benchmark_dataset(Species.human, Modification.psi, True)
train_dataset = load_benchmark_dataset(Species.human, Modification.psi)
```

```python
def replace_middle_char(s, new_char):
    # Calculate the middle index
    mid = len(s) // 2
    
    # For odd-length strings
    if len(s) % 2 != 0:
        return s[:mid] + new_char + s[mid+1:]
    
    # For even-length strings, replace the character just before the midpoint
    else:
        return s[:mid-1] + new_char + s[mid:]
```

```python
for sample in train_dataset.samples.values:
    sf = sample[0]
    sr = sf[::-1]
    
    s1, c1 = RNA.fold(sf)
    s2, c2 = RNA.fold(sr)
    
    s11, c11 = RNA.fold(replace_middle_char(sf, 'P'))
    s22, c22 = RNA.fold(replace_middle_char(sr, 'P'))
    
    print(s1, s2, f'({c1}, {c2})')
    print(s11, s22, f'({c11}, {c22})')
    print()
```
